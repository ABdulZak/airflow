```python
def _clean_data(filepath=''):
    def clean_column(id, lst):
        clean_df = pd.DataFrame()
        for i in range(len(lst)):
            clean_df = pd.concat([clean_df, pd.DataFrame(lst[i], index=[id])])
            clean_df['id'] = id
        return clean_df

    def extract_json(df):
        clean_df = pd.DataFrame()
        for i in range(len(df)):
            if pd.isnull(df.iloc[i,1]):
                continue
            lst= df.iloc[i,1]
            extract_json = re.findall(r'\{.*?\}', lst)  
            lst = [json.loads(i) for i in extract_json]
            clean_df = pd.concat([clean_df, clean_column(df.iloc[i,0], lst)], ignore_index=True)
        return clean_df
    

    df = pd.read_csv(filepath)
    df['id'] = df['id'].astype(int)
    contracts = df[['id', 'contracts']]

    cleaned_contracts = extract_json(contracts)
    cleaned_contracts = cleaned_contracts[['id','contract_id','bank','summa','loan_summa','claim_date','claim_id','contract_date']]

    df.drop('contracts', axis=1, inplace=True)

    cleaned_contracts.to_csv('cleaned_contracts.csv', index=False)
    df.to_csv('cleaned_data.csv', index=False)

_clean_data('../../include/data.csv')
```
